```python
from faker import Faker
import requests
import random
import time

events = 5000

pages = [
    "/home", "/about", "/contact", "/pricing", "/blog",
    "/faq", "/login", "/signup", "/dashboard", "/settings"
]

fake = Faker()

session_ids = [fake.uuid4() for _ in range(20)]

referrers = [
    "https://www.google.com",
    "https://www.facebook.com",
    "https://www.twitter.com",
    "https://www.linkedin.com",
    "https://www.reddit.com",
]

base_url = "http://localhost:8002"
create_endpoint = f"{base_url}/api/events/"

success = 0
failed = 0

start = time.time()

# 🔥 IMPORTANT: reuse connection
with requests.Session() as client:
    for i in range(events):

        payload = {
            "page": random.choice(pages),
            "user_agent": fake.user_agent(),
            "ip_address": fake.ipv4(),
            "duration": random.randint(1, 300),
            "session_id": random.choice(session_ids),
            "referrer": random.choice(referrers),
        }

        try:
            response = client.post(create_endpoint, json=payload, timeout=10)

            if response.ok:
                success += 1
            else:
                failed += 1

        except Exception:
            failed += 1

        # print less (VERY IMPORTANT)
        if i % 500 == 0:
            print(f"Progress: {i}/{events}")

end = time.time()

print("DONE")
print("Success:", success)
print("Failed:", failed)
print("Time:", round(end - start, 2), "seconds")
```
